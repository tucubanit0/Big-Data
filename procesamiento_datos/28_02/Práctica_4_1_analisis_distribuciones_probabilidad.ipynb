#### Actividad Práctica 4.1: Análisis de Distribuciones de Probabilidad

-Objetivo
El objetivo es analizar una distribución de probabilidad en un conjunto de datos utilizando herramientas estadísticas en Python.

Paso 1: Selección del conjunto de datos
Para este ejercicio, elegiremos datos simulados de alturas de personas en centímetros, que es una variable numérica continua.

 -Justificación:
 La altura es un ejemplo típico de una variable que sigue aproximadamente una distribución normal en poblaciones grandes. Esto permitirá explorar herramientas estadísticas y ajustar una distribución teórica.



Paso 2: Generar y cargar los datos
En este caso, generaremos datos simulados utilizando una distribución normal con:
Media = 170 cm
Desviación estándar = 10 cm
Tamaño de la muestra = 500
Estos valores son razonables para una población adulta.

```python
import numpy as np        # Importamos numpy que es una librería para trabajar con vectores y matrices
import pandas as pd       # Importamos pandas que es una librería para trabajar con estructuras de datos


# Fijamos la semilla para reproducibilidad
np.random.seed(42)

# Generamos datos de alturas simulados con distribución normal
alturas = np.random.normal(loc=170, scale=10, size=500)

# Convertimos en un DataFrame
df = pd.DataFrame(alturas, columns=["Altura"])

# Mostramos las primeras filas del dataset
df.head()
```

-Explicación:
np.random.normal(loc=170, scale=10, size=500): Genera 500 datos siguiendo una distribución normal con media 170 y desviación estándar 10.
Se almacenan en un DataFrame para facilitar el análisis.


Paso 3: Análisis estadístico
Ahora calculamos las principales métricas estadísticas:
Media (promedio): Valor central de la distribución.
Varianza: Cuantifica la dispersión de los datos.
Skewness (asimetría): Mide si la distribución es simétrica o sesgada.


```python
'''Análisis estadístico
Ahora calculamos las principales métricas estadísticas:'''

#Media (promedio): Valor central de la distribución.
media = df["Altura"].mean()                # Esta función calcula la media de la columna Altura
print(f"Media: {media:.2f}")               # Mostramos la media con dos decimales usando media:.2f en donde .2f indica que se mostrarán dos decimales

# Varianza: Cuantifica la dispersión de los datos.
varianza = df["Altura"].var()              # Esta función calcula la varianza de la columna Altura
print(f"Varianza: {varianza:.2f}")         # Mostramos la varianza con dos decimales usando varianza:.2f en donde .2f indica que se mostrarán dos decimales

# Skewness (asimetría): Mide si la distribución es simétrica o sesgada.
skewness = df["Altura"].skew()             # Esta función calcula la asimetría de la columna Altura usando el método de Pearson para calcular la asimetría que es .skew()
print(f"Asimetría: {skewness:.2f}")        # Mostramos la asimetría con dos decimales usando skewness:.2f en donde .2f indica que se mostrarán dos decimales
```

Explicación:

Paso 4: Visualización con Histograma
Para comprender la distribución, creamos un histograma y superponemos una curva de densidad.

```python
# Visualización con Histograma para comprender la distribución, creamos un histograma y superponemos una curva de densidad.
import matplotlib.pyplot as plt    # Importamos la librería matplotlib que será la encargada de realizar la visualización de los datos.

plt.figure(figsize=(10, 6))         # Creamos una figura de tamaño 10x6
plt.hist(df["Altura"], bins=30, density=True, alpha=0.6, color='g')  # Creamos un histograma con 30 bins, densidad y transparencia de 0.6
plt.title("Histograma de alturas")           # Título del gráfico
plt.xlabel("Altura")                      # Etiqueta del eje x
plt.ylabel("Frecuencia")                    # Etiqueta del eje y
plt.grid(True)                        # Mostramos la grilla, que la grilla es una cuadrícula que se dibuja en el gráfico
df["Altura"].plot(kind='kde', color='b')  # Creamos una curva de densidad de la columna Altura, en donde kind='kde' indica que se creará una curva de densidad y color='b' indica que será de color azul

plt.legend(["Densidad", "Altura"])        # Mostramos la leyenda con los nombres "Densidad" y "Altura"
plt.show()
```

Explicación:

Paso 5: Ajuste de una distribución teórica
Ahora ajustamos una distribución normal y la superponemos en el histograma.

```python
# Ajuste de una distribución teórica. Ahora ajustamos una distribución normal y la superponemos en el histograma.
from scipy.stats import norm       # Importamos from scipy.stats import norm que es una librería para trabajar con distribuciones estadísticas.

# Ajustamos una distribución normal
mu, std = norm.fit(df["Altura"])         # Ajustamos una distribución normal a la columna Altura usando norm.fit() que nos devuelve la media y la desviación estándar

plt.figure(figsize=(10, 6))         # Creamos una figura de tamaño 10x6
plt.hist(df["Altura"], bins=30, density=True, alpha=0.6, color='g') # Creamos un histograma con 30 bins, densidad y transparencia de 0.6
plt.title("Histograma de alturas") # Título del gráfico
plt.xlabel("Altura")               # Etiqueta del eje x
plt.ylabel("Frecuencia")           # Etiqueta del eje y
plt.grid(True)                    # Mostramos la grilla, utilizamos plt.grid(True) para mostrar la grilla, si no queremos mostrar la grilla, utilizamos plt.grid(False)
xmin, xmax = plt.xlim()           # Obtenemos los valores mínimos y máximos del eje x
x = np.linspace(xmin, xmax, 100)  # Creamos un vector de 100 puntos entre xmin y xmax
p = norm.pdf(x, mu, std)          # Calculamos la función de densidad de probabilidad de la distribución normal, std es la desviación estándar y mu es la media
plt.plot(x, p, 'k', linewidth=2) # Graficamos la distribución normal con color, 'k' indica que será de color y linewidth=2 indica que el grosor de la línea será de 2

plt.legend(["Distribución normal", "Altura"]) # Mostramos la leyenda con los nombres "Distribución normal" y "Altura"
plt.show()
```

```python
# Interpretación de Resultados. Los resultados obtenidos fueron:
# Media: 169.98
mean_result = 169.98
print(f"Media: {mean_result}")

# Varianza: 101.63
variance_result = 101.63
print(f"Varianza: {variance_result}")

# Skewness(asimetría): 0.02
skewness_result = 0.02
print(f"Asimetría: {skewness_result}")

```
